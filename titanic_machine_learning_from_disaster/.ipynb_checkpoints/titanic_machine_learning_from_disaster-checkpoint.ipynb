```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
%matplotlib inline
font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 16}

rc('font', **font)
```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

```python
print("Number of Null Values in Each Column:\n")
for col_name in df_train.columns:
    print("%s: %s"%(col_name, df_train[col_name].isnull().sum()))
```

```python
df_train['Sex'] = df_train['Sex'].astype('category').cat.codes

fig, axes = plt.subplots(4, 3,figsize=(12, 12), facecolor='w', edgecolor='k')#, subplot_kw=dict(polar=True))

kde_cols = ['Age','Fare','Pclass','Sex']
colors = ['crimson','limegreen']
labels = ['Not Survived','Survived']
nbins = [20,20,3,2]
df_list = [df_train,
            df_train[df_train['Age'].notnull()],
            df_train[df_train['Age'].isnull()]]
title_list = ['Training Set, All Data Points', 'Training Set, Known Age','Training Set, Missing Age']

for col_name in kde_cols:
     for i in [0,1,2]:
        df_plot = df_list[i]
        for j in [0,1]:
            ax = df_plot[(df_plot['Survived']==j)][col_name].plot(ax=axes[kde_cols.index(col_name),i],kind='hist', 
                                                             alpha=0.7,color=colors[j], edgecolor = 'black',
                                                             #histtype='step',
                                                             lw=1, label=labels[j],
                                                            bins=nbins[kde_cols.index(col_name)],sharey=True)
            if col_name=='Age': ax.set_title(title_list[i])#,fontsize=22)
            ax.set_xlabel(col_name)#,fontsize=16)
            if i > 0: ax.set_ylabel("")
            
            if kde_cols.index(col_name) == 0: ax.set_ylim([0,100])


fig.subplots_adjust(bottom=0.3, wspace=0.33)
fig.legend(ncol=1,loc=1,bbox_to_anchor=(0.96,0.96),fancybox=False, shadow=False)
fig.tight_layout()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor

features =  ['Survived', 'Pclass', 'Sex']#, 'Fare']
X_train = df_train[df_train['Age'].notnull()].loc[:,features]
X_train['Sex'] = X_train['Sex'].astype('category').cat.codes

label = ['Age']
y_train = df_train[df_train['Age'].notnull()].loc[:,label]
y_train


features =  ['Survived', 'Pclass', 'Sex']#, 'Fare']
X_test = df_train[df_train['Age'].isnull()].loc[:,features]
X_test['Sex'] = X_test['Sex'].astype('category').cat.codes
X_test

from sklearn.neural_network import MLPRegressor
mlp = MLPRegressor(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(3,3,3),random_state=1, max_iter=10000, shuffle=True)
mlp.fit(X=X_train,y=y_train)
```

```python
mlp.score(X=X_train,y=y_train)
```

```python
predictions = mlp.predict(X_test)
```

```python
%matplotlib inline
plt.hist(predictions)
```
